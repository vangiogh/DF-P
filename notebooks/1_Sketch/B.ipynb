# Sketch data processing

## Load data

```python
import pandas as pd

df = pd.read_csv('../../data/gapminder.csv')
df
```

## Filter data

### Masking

```python
year = 2007

mask_year = df['year'] == year
df_year = df[mask_year].copy()

df_year
```

### Visualization

```python
import plotly.express as px

fig = px.scatter(df_year, x='gdpPercap', y='lifeExp', size='pop', color='continent', hover_name='country', log_x=True, size_max=60)
fig
```
